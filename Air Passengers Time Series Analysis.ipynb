```python
import os
import pandas as pd


kaggle_json = '{"username":"....","key":"...."}'


kaggle_dir = os.path.expanduser("~/.kaggle")
os.makedirs(kaggle_dir, exist_ok=True)
kaggle_file_path = os.path.join(kaggle_dir, "kaggle.json")


with open(kaggle_file_path, "w") as f:
    f.write(kaggle_json)

os.chmod(kaggle_file_path, 0o600)

datasets = {
    "rakannimer/air-passengers": "Air_Passengers",
    "sumanthvrao/daily-climate-time-series-data": "Daily_Climate",
    "sudalairajkumar/daily-temperature-of-major-cities": "Daily_Temperature",
    "rohanrao/nifty50-stock-market-data": "Stock_Market",
}

for dataset_slug, output_dir in datasets.items():
    print(f"Downloading {dataset_slug}...")
    os.system(f"kaggle datasets download -d {dataset_slug} -p {output_dir}")
    os.system(f"unzip -q {output_dir}/*.zip -d {output_dir}")
    print(f"{dataset_slug} downloaded and extracted successfully!\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1.columns
```

```python
df1['Month'] = pd.to_datetime(df1['Month'])
df1 = df1.sort_values(by='Month')
```

```python
df1.set_index('Month', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df1['#Passengers'], label="Passengers", color='blue', linewidth=2)
plt.title('Monthly Air Passengers Over Time (Line Chart)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
fig1 = go.Figure()

fig1.add_trace(go.Scatter(x=df1['Month'], y=df1['#Passengers'], mode='lines', name='Passengers'))

fig1.update_layout(
    title='Monthly Air Passengers Over Time',
    xaxis_title='Month',
    yaxis_title='Number of Passengers',
    template='plotly_dark',
    xaxis_rangeslider_visible=True
)

fig1.show()
```
